```python
#створіть функцію для обчислення функції втрат у векторному вигляді
def cost(y: np.array, h: np.array) -> float:
    return np.mean((h - y) ** 2) / 2
```

```python
class LinearRegression:
    def __init__(self, lr: float = 0.0001, thr: float = 0.00001, n_epochs: int = 1000):
        self.lr = lr
        self.thr = thr
        self.n_epochs = n_epochs
        self.w = np.array([random.uniform(-5, 5), random.uniform(-5, 5)])

    #напишіть функцію гіпотези лінійної регресії у векторному вигляді
    def predict(self, X: np.ndarray) -> list:
        return X @ self.w

    #реалізуйте один крок градієнтного спуску
    def update_w(self, X: np.array, y: np.array):
        # w = w - lr / m * XT @ (w @ X - y)
        m = len(y)

        h = self.predict(X)

        self.w -= self.lr / m * X.T @ (h - y)

    def fit(self, X: np.array, y: np.array):
        last_cost = 1000000

        for i in range(self.n_epochs):
            self.update_w(X, y)
            new_cost = cost(y, self.predict(X))
            print(f"{i}, {self.w}, {new_cost}")
            if last_cost - new_cost < self.thr:
                break
            last_cost = new_cost
```

```python
class LinearRegression:
    def __init__(self, lr: float = 0.0001, thr: float = 0.00001, n_epochs: int = 1000):
        self.lr = lr
        self.thr = thr
        self.n_epochs = n_epochs
        self.w = np.array([random.uniform(-1, 1), random.uniform(-1, 1), random.uniform(-1, 1)])

    #напишіть функцію гіпотези лінійної регресії у векторному вигляді
    def predict(self, X: np.ndarray) -> list:
        return X @ self.w

    #реалізуйте один крок градієнтного спуску
    def update_w(self, X: np.array, y: np.array):
        # w = w - lr / m * XT @ (w @ X - y)
        m = len(y)

        h = self.predict(X)

        self.w -= self.lr / m * X.T @ (h - y)

    def fit(self, X: np.array, y: np.array):
        last_cost = 1000000

        for i in range(self.n_epochs):
            self.update_w(X, y)
            new_cost = cost(y, self.predict(X))
            print(f"{i}, {self.w}, {new_cost}")
            if last_cost == new_cost:
                break
            last_cost = new_cost
```

```python
X = np.array(data[['area', 'bedrooms', 'bathrooms']])
y = np.array(data['price'])

X_normalized = (X - X.mean(axis=0)) / X.std(axis=0)
Y_normalized = (y - y.mean(axis=0)) / y.std(axis=0)

regression_model = LinearRegression(lr=0.01, n_epochs=10000, thr=0.0000001)
regression_model.fit(X_normalized, Y_normalized)

# Виведення найкращих параметрів w
regression_model.w
```
